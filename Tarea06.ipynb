```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10,0.1)
y=3*np.sin(x)+4*np.cos(2*x)
plt.plot(x,y, color="green", linewidth=2)
plt.title("3Sen(x)+ 4Cos(2x)")
plt.xlabel("Posici√≥n (m)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(1000)
y = np.random.rand(1000)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=20, alpha=0.4, color='purple')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Gr√°fica de Dispersi√≥n de 1000 Puntos Aleatorios')
plt.grid(True)
plt.show()
```
