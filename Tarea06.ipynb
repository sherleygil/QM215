```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10,0.1)
y=3*np.sin(x)+4*np.cos(2*x)
plt.plot(x,y, color="green", linewidth=2)
plt.title("3Sen(x)+ 4Cos(2x)")
plt.xlabel("Posición (m)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(1000)
y = np.random.rand(1000)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=20, alpha=0.4, color='purple')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Gráfica de Dispersión de 1000 Puntos Aleatorios')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
categorias=["Adm. de Empresas y Contabilidad", "Adm. Pública", "Arquitectura","Ciencia de la Eduación", "Ciencias Naturales y Exactas", "Comunicación Social", "Derecho y Ciencias Políticas", "Economía","Enfermería", "Humanidades", "Medicina"]
valores=[1412, 1142, 324, 2238, 1342, 546, 1910, 657, 404, 1783, 1047]
plt.bar(categorias, valores, color='mediumseagreen', edgecolor='black')
plt.title("Número de Estudiantes Por cada Facultad", fontsize=14)
plt.xlabel("Facultad", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.ylabel("Número de Estudiantes", fontsize=12)
plt.grid(True)
plt.show()
```
